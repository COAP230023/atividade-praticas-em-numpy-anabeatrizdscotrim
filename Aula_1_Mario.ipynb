```python
from PIL import Image, ImageChops
import matplotlib.pyplot as plt
import requests
import io

url = 'https://images.unsplash.com/photo-1759171053256-bf1b480a310a?ixlib=rb-4.1.0&q=85&fm=jpg&crop=entropy&cs=srgb&w=640'

image = Image.open(requests.get(url, stream=True).raw)
image = image.convert("RGB")

r, g, b = image.split()

#2. Imagem com Cores Invertidas (Negativo)
# Inverte os valores de cada pixel (p para 255 - p)
inverted_image = ImageChops.invert(image)
r_inv, g_inv, b_inv = inverted_image.split()

# O m√©todo 'L' converte a imagem para Escala de Cinza (Luminance)
gray_image = image.convert('L')

colors = {
    'R': 'Reds',
    'G': 'Greens',
    'B': 'Blues'
}

fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for ax, channel, color in zip(axs[0], [r, g, b], ['R','G','B']):
    ax.imshow(channel, cmap=colors[color])
    ax.set_title(f'Canal {color} Original')
    ax.axis('off')

for ax, channel, color in zip(axs[1], [r_inv, g_inv, b_inv], ['R','G','B']):
    ax.imshow(channel, cmap=colors[color])
    ax.set_title(f'Canal {color} Negativo')
    ax.axis('off')

for i, ax in enumerate(axs[2]):
    ax.imshow(gray_image, cmap='gray')
    if i == 1:
        ax.set_title('Imagem em Tons de Cinza')
    else:
        ax.set_title('Tons de Cinza')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
